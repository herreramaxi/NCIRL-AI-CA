# Datasets

## Clasification task

### 200+ Financial Indicators of US stocks (2014-2018)

<b>Source:</b> [200-financial-indicators-of-us-stocks-20142018](https://www.kaggle.com/cnic92/200-financial-indicators-of-us-stocks-20142018)<br>
<b>File:</b>  [lassification/2018_Financial_Data.csv](https://github.com/herreramaxi/NCIRL-AI-CA/blob/main/classification/2018_Financial_Data.csv)<br>
<b>Description:</b>  The dataset contains more than 200 financial indicators, for around 4k US stocks. <br>
<b>Shape:</b>  4392 rows × 225 columns.<br>
The dataset is composed by 225 columns, the first column is the symbol/ticker, and the rest are financial indicators.<br>
The last two columns are the labeled target columns:<br>
•	2019 PRICE VAR [%]: Percent price variation.<br>
•	Class: Binary attribute that indicates with “1” if a trader should buy the stock at the beginning of the year and sell at the end for a profit. A “0” indicates that a trader should not buy it as it would produce loss of capital.


## Regression task

### Prices

This is a subset of Wiki prices with a starting date of 2000 until 2018.<br>
<b>File:</b> [regression/prices-2000-2018.csv](https://github.com/herreramaxi/NCIRL-AI-CA/blob/main/regression/prices-2000-2018.csv)
<b>Generated by:</b> [generateDatasets.ipynb](https://github.com/herreramaxi/NCIRL-AI-CA/blob/main/generateDatasets.ipynb)<br>
<b>Source:</b>
* https://data.nasdaq.com/tables/WIKIP/WIKI-PRICES/
* https://data.nasdaq.com/databases/WIKIP/documentation

<b>Description:</b> End of day stock prices, dividends and splits for 3,000 US companies, curated by the Quandl community and released into the public domain.History: 1996 -2018<br>
<b>Shape:</b> 4706 rows × 3199 columns.<br>
The dataset is composed by 3199 columns, the first column is the date, and the rest are tickers/symbols. The cell values are Adjusted Closing Price
